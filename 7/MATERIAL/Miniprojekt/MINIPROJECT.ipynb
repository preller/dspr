5. Next, perform a clustering using the optimal number of clusters. Also, visualize your results in an appropriate fashion.


```python
# By visualy analyzing the data, there seems to be 5 clusters in the most interesting plot (income vs score).
# Hower, just to be sure, let's use the elbow method to find the optimal number of clusters:

dataFiltered = data[['Annual Income (k$)', 'Spending Score (1-100)']]

distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='k-means++', # this can be left out, as it is the default
                n_init=10, 
                max_iter=300, 
                random_state=0)
    km.fit(dataFiltered)
    distortions.append(km.inertia_)
plt.plot(range(1,11), distortions , marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
#plt.savefig('./figures/elbow.png', dpi=300)
plt.show()

#TODO CHECK WHY NOT 3 or 4! https://stackoverflow.com/questions/19197715/scikit-learn-k-means-elbow-criterion
```

```python
# And the elbow seems to be when k = 5, as we initially thought. Let's use that value:
X = dataFiltered
CLUSTERS = 5
X
km = KMeans(n_clusters=CLUSTERS, 
            init='random', # leave this out for k-means++ (see below)
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(dataFiltered)
print(y_km)
# # And now that we have the clusters, let's plot them:
plt.scatter(X[y_km==0], 
            X[y_km==0], #//TODO DEBUG THIS
            s=50, 
            c='lightgreen', 
            marker='s', 
            label='cluster 1')
plt.scatter(X[y_km==1], 
            X[y_km==1], 
            s=50, 
            c='lightblue', 
            marker='s', 
            label='cluster 1')
plt.scatter(km.cluster_centers_[:,0], 
            km.cluster_centers_[:,1], 
            s=250, 
            marker='*', 
            c='red', 
            label='centroids')
plt.legend()
plt.grid()
plt.tight_layout()
#plt.savefig('./figures/centroids.png', dpi=300)
plt.show()
```
